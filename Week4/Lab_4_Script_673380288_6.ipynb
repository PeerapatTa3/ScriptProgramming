# lab4_1.py - Student Management System (List Operations)

```python
# lab4_1.py - Student Management System (List Operations)
# หมายเหตุ: โค้ดนี้ตัดส่วน "Challenge" (ระบบเมนูแบบ while loop) ออก
#          เพื่อให้นักศึกษานำ concept ด้านล่างไปต่อยอดคิดเองเป็นแบบฝึกหัด

print("--- Lab 4.1: Student Management System ---")

# 1) เริ่มต้นด้วย list ว่างสำหรับเก็บชื่อนักศึกษา
student_names = []

# 2) เพิ่มนักศึกษา: รับชื่อจากผู้ใช้ 3 คน แล้ว append() เข้า list
def add_student():
  print("\n[1] Add Students")
  for i in range(3): #กำหนดตัวแปรใน loop 👈ต้องทำ
      st = input(f"Enter student name #{i + 1}: ") #กำหนดตัวแปร 👈ต้องทำ
      student_names.append(st)

  print(f"Current list: {student_names}")

# 3) ค้นหานักศึกษา: ใช้ 'in' เพื่อเช็คว่ามีชื่อนี้อยู่ใน list หรือไม่
#    ถ้าพบ ให้แสดง index ด้วย index()
def search_student():
  print("\n[2] Find a Student")
  search_name = input("Enter a name to search for: ")

  if search_name in student_names: #เขียนเงื่อนไข ถ้าหากมีชื่อในลิส: 👈ต้องทำ
      position = student_names.index(search_name)
      print(f"Found '{search_name}' at index {position}")
  else:
      print(f"'{search_name}' not found in the list")

# 4) ลบนักศึกษา: เช็คก่อนว่ามีอยู่จริงหรือไม่ (กัน error) แล้วค่อย remove()
def remove_student():
  print("\n[3] Remove a Student")
  remove_name = input("Enter a name to remove: ")

  if remove_name in student_names:
      student_names.remove(remove_name) #เติมฟังก์ชัน remove(remove_name) 👈ต้องทำ
      print(f"'{remove_name}' has been removed")
  else:
      print(f"'{remove_name}' not found, nothing removed")

  print(f"Updated list: {student_names}")

# 5) เรียงลำดับรายชื่อตามตัวอักษรด้วย sort()
def sort_student():
  print("\n[4] Sort Students")
  student_names.sort() #เติมฟังก์ชันsort()
  print(f"Sorted list: {student_names}")

# 6) นับจำนวนนักศึกษาทั้งหมดด้วย len()w
def count_student():
  print("\n[5] Count Students")
  print(f"Total students: {len(student_names)}") #ใช้คำสั่งlen() หาจำนวนนักศึกษาในลิส student_names👈ต้องทำ

# --------------------------------------------------------------------
# ให้นักศึกษาลองต่อยอดเอง (Challenge):
# ลองเขียนระบบเมนูด้วย while loop (จาก Week 3) ที่ให้ผู้ใช้เลือกได้ว่า12
# จะเพิ่ม / ลบ / ค้นหา / เรียงลำดับ / หรือออกจากโปรแกรม
# โดยนำโค้ดแต่ละส่วนด้านบนไปจัดเป็นฟังก์ชันหรือเงื่อนไขในเมนู
# --------------------------------------------------------------------

def get_choice():
  print("\nMenu:")
  print("1. Add Students")
  print("2. Remove a Student")
  print("3. Find a Student")
  print("4. Sort Students")
  print("5. Count Students")
  print("6. Exit")

  return  input("Enter your choice : ")

while True:
  user_choice = get_choice()

  match user_choice:
    case "1":
      add_student()
    case "2":
      remove_student()
    case "3":
      search_student()
    case "4":
      sort_student()
    case "5":
      count_student()
    case "6":
      print("Exited menu.")
      break
    case _:
      print("Invalid choice. Please try again.")
```

# lab4_2.py - Geometric Calculations (Tuples & Immutability)

```python
from re import X
# lab4_2.py - Geometric Calculations (Tuples & Immutability)
# หมายเหตุ: โค้ดนี้ตัดส่วน "Challenge" (รับพิกัดจากผู้ใช้เพื่อคำนวณระยะทาง) ออก
#          เพื่อให้นักศึกษานำ concept ด้านล่างไปต่อยอดคิดเองเป็นแบบฝึกหัด

import math

def get_coordinate():
  try:
    point = input(f"Input coordinate x,y : ")
    x, y = map(float, point.split(","))
  except:
    print("Input invalid.")
  return (x,y)

print("--- Lab 4.2: Geometric Calculations ---")

# 1) กำหนดจุดสองจุดด้วย tuple: point1 และ point2
point1 = get_coordinate() #เติมพิกัด x y ของจุด 👈ต้องทำ
point2 = get_coordinate() #เติมพิกัด x y ของจุด 👈ต้องทำ

# 2) เข้าถึงค่าพิกัด x, y ของแต่ละจุด
print("\n[1] Access Coordinates")
print(f"Point 1: x = {point1[0]}, y = {point1[1]}") #เติมเลข 0 และ 1 ใน [] 👈ต้องทำ
print(f"Point 2: x = {point2[0]}, y = {point2[1]}") #เติมเลข 0 และ 1 ใน [] 👈ต้องทำ

# 3) ทดลองแก้ไขค่าใน tuple เพื่อแสดงให้เห็นว่า tuple เปลี่ยนแปลงไม่ได้ (immutable)
print("\n[2] Attempt Modification (Immutability Demonstration)")
try:
    point1[0] = 5  # บรรทัดนี้จะทำให้เกิด TypeError เพราะ tuple แก้ไขไม่ได้
except TypeError as e:
    print(f"Error trying to modify tuple: {e}")

# 4) คำนวณระยะห่างระหว่างจุดสองจุดด้วยสูตรระยะทาง
#    distance = sqrt( (x2 - x1)^2 + (y2 - y1)^2 )
print("\n[3] Calculate Distance")
distance = math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2) #เติมเลขตำแหน่งในลิส 0 และ 1 ใน [] ให้ถูกต้อง 👈ต้องทำ
print(f"Distance between point1 {point1} and point2 {point2}: {distance}")

# --------------------------------------------------------------------
# ให้นักศึกษาลองต่อยอดเอง (Challenge):
# ลองแก้โค้ดให้รับพิกัดของจุดใหม่ 2 จุดจากผู้ใช้ผ่าน input()
# (อย่าลืมแปลงชนิดข้อมูลจาก input() ให้เป็นตัวเลขก่อนนำไปคำนวณ)
# แล้วคำนวณระยะห่างระหว่างจุดทั้งสองที่ผู้ใช้กรอกเข้ามา
# --------------------------------------------------------------------
```

#Activity 4.2: Interactive To-Do List (using List Operations)
สิ่งที่อยากให้ทำ : ให้นักเรียนพัฒนาโปรแกรมจัดการรายการสิ่งที่ต้องทำ (To-Do List) แบบข้อความ (Text-based) อย่างง่าย โดยประยุกต์ใช้ความรู้เรื่องการจัดการ List ที่ได้เรียนมา

#สิ่งที่ต้องมี

1. สร้าง List ว่างเริ่มต้นไว้ตัวหนึ่ง ชื่อว่า todo_list

2. ใช้การวนลูป while เพื่อแสดงเมนูคำสั่งให้ผู้ใช้เลือก ดังนี้:

>2.1 เพิ่มรายการงาน (Add a task)
>
>2.2 ดูรายการงานทั้งหมด (View tasks)
>
>2.3 ทำเครื่องหมายว่างานเสร็จสิ้นแล้ว (ลบงานออก โดยเลือกจากตำแหน่ง Index
หรือตามชื่อคำสั่ง)
>
>2.4 ออกจากโปรแกรม (Exit)

3. ใช้คำสั่ง input() เพื่อรับค่าเมนูที่ผู้ใช้เลือก

4. ใช้คำสั่งจัดการ List ต่างๆ (เช่น append(), remove() หรือ pop(), len()) ในการบริหารจัดการข้อมูลใน todo_list

```python
todo_list = []

def get_choice():
  print("\nMenu:")
  print("1. Add a task")
  print("2. View tasks")
  print("3. Remove a task")
  print("4. Exit")

  return  input("Enter your choice : ")

def add_task():
  task = input("New task : ")
  todo_list.append(task)
  print(f"{task} added to the list!")

def view_tasks():
  if todo_list:
    print("Tasks:")
    for task in todo_list:
      print(task)
  else:
    print("No tasks found.")

def remove_task():
  to_remove = input("Task to remove : ")
  try:
    to_remove = int(to_remove)
    if 0 <= to_remove < len(todo_list):
      del todo_list[to_remove]
      print("Removed task!")
    else:
      print("Invalid index!")
  except:
    if to_remove in todo_list:
      todo_list.remove(to_remove)
      print("Removed task!")
    else:
      print("Task not found!")

while True:
  user_choice = get_choice()

  match user_choice:
    case "1":
      add_task()
    case "2":
      view_tasks()
    case "3":
      remove_task()
    case "4":
      break
      print("Exited menu.")
    case _:
      print("Invalid choice. Please try again.")
```
